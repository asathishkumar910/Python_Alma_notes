````python
#Good Example
temperature_celsius = 25
````

````python
my_variable = 42
print(My_Variable) #would throw an error
````

**Snake Case:** The recommended convention for naming variables is to use snake_case, where words are separated by underscores. This makes the variable name more readable.

eg: snake_case_variable

**Statement -1:** Lists in Python are ordered collections of elements, and the elements in a list must be of the same data type.
True or False?

**Statement-2:** Lists are mutable, meaning you can change their elements after creating the list.
True or False?

**Statement-3:** When you modify a list that is assigned to another variable, both the original list and the new variable will reflect the changes, as they point to the same list object in memory.
True or False?

**Statement-4:** The 'type()' function in Python can be used to determine what is the datatype of a variable.
True or False?

**Statement-5:** The 'len()' function and the 'count()' method both can be used to find the number of elements in a list.
True or False?

#### **Answers:**

* Statement 1: False. Python lists are indeed ordered collections, meaning the order of elements is maintained. However, the elements in a list can be of different data types. A single list can contain integers, strings, objects, or even other lists.

* Statement 2: True. In Python, lists are mutable, which means you can modify their content after creation. You can add, remove, or change elements in a list.

##Introduction to Tuples


In Python, tuples are similar to lists in many ways, but they have some important differences that make them useful in certain situations. One of the key differences is that tuples are immutable, which means that you can't modify them after they've been created. üîí

This might seem like a limitation, but it can actually be quite useful in certain scenarios. For example, if you have a set of data that you know won't change, you can store it in a tuple to make sure that it stays the same. This can be particularly important if you're working on a large project with many different pieces of code, because it helps to ensure that your data stays consistent throughout.

In brief:

* In Python tuples are very similar to lists, however, unlike lists they are *immutable* meaning they can not be changed.

###Mutability



In Python, dictionaries are mutable, which means that you can modify their contents after they've been created. This makes them a very flexible and powerful data structure for working with complex data. üîß

When you modify a dictionary, you can add, remove, or update key-value pairs as needed. For example, you might add a new key-value pair to represent a new piece of data, or update an existing key-value pair to reflect a change in the data. You can also remove key-value pairs that are no longer needed.

Here's an example of modifying a dictionary:

````python
customer = {
    "id": 1234,
    "name": "John Smith",
    "email": "john@example.com"
}

# Update the customer's email address
customer["email"] = "john.smith@example.com"

# Add a new key-value pair
customer["phone"] = "555-555-1212"

# Remove the customer's ID
del customer["id"]

print(customer)
````

As you can see, the changes we made to the dictionary have been reflected in the output. This is the power of mutability in Python dictionaries: it allows us to modify data structures in place, without having to create entirely new data structures from scratch.

## Ordered and Unordered Data Types in Python

In Python, data types can be broadly categorized into two groups: ordered and unordered data types. These categories define how the elements in the data type are stored and accessed.

###Ordered Data Types:

Ordered data types maintain the sequence or order of elements. You can access elements by their position, and they are **indexed** with integers, starting from 0. You'll be covering indexing in upcoming lessons in detail. Common examples of ordered data types in Python include strings, lists, and tuples.

**String (str)**: A string is a sequence of characters, and each character has a specific position within the string. You can access individual characters or substrings by their index.

````python
text = "Hello, Python!"
print(text[0])  # Output: 'H'
print(text[7:13])  # Output: 'Python'
````

**List (list)**: Lists are ordered collections of items. You can access list elements by their index, and they maintain the order in which items were added.

````python
fruits = ['apple', 'banana', 'cherry']
print(fruits[1])  # Output: 'banana'
````

Here's a question for you. Do you think ['apple', 'banana', 'cherry'] and  ['apple','cherry', 'banana'] are one and the same?

Logically thinking, they should be right?

Let's see what python thinks:

````python
fruits = ['apple', 'banana', 'cherry']
fruits1= ['apple', 'cherry','banana']

fruits==fruits1  #Using comparison operator to check if two variables are equal.
````

Python clearly does not agree. This is so because list is an ordered data type.

Notice how the two variables became equal when we change the order:

````python
fruits = ['apple', 'banana', 'cherry']
fruits1= ['apple', 'banana', 'cherry']

fruits==fruits1  #Using comparison operator to check if two variables are equal.
````

**Tuple (tuple):** Tuples are similar to lists, but remember they are immutable, meaning their elements cannot be changed after creation. They also maintain order.

````python
coordinates = (3, 4)
print(coordinates[0])  # Output: 3
````

###Unordered Data Types:

Unordered data types do not maintain the order of elements, and you cannot access elements by their position. Common examples of unordered data types in Python include sets and dictionaries.

**Set (set)**: A set is an unordered collection of unique elements. Sets are commonly used for membership testing or eliminating duplicates.

````python
colors = {"red", "green", "blue"}
````

````python
colors[0]  #Would throw you an error
````

Since sets are unordered, could it be the case the order of the elements in the set would not matter when we compare two sets? ü§î.
Let's find out

````python
colors = {"red", "green", "blue"}
colors1={"green","red", "blue"}
colors==colors1
````

Voila! It really is the case!!

**Dictionary (dict)**: Dictionaries as you know are collections of key-value pairs. They are not ordered, but you access values by their associated keys.

````python
person = {"name": "John", "age": 30}
print(person["name"])  # Output: 'John'
````

In summary, understanding whether a data type is ordered or unordered is crucial when working with Python, as it affects how you can manipulate and access the data within these data structures. Ordered data types are suitable when maintaining the order is important, while unordered data types are useful for quick lookups, uniqueness checks, or organizing data based on keys.

###**Best Practices/Tips**
Best Practices/Tips to remember

1. Use descriptive variable names: When naming your variables, use descriptive phrases that indicate the data type the variable will hold. For example, use "age" instead of "a" for a variable that will have an age value.
2. Use the appropriate data types: Choose the proper data type for your variables and objects to ensure data accuracy and optimize code performance. For example, use integers instead of floating-point numbers for mathematical operations and lists instead of dictionaries for iteration and indexing.
3. Convert data types when necessary: Sometimes, you may need to convert data from one type to another. For example, you may need to convert a string to an integer or a float to an integer. Use built-in Python functions like int(), float(), and str() to convert between data types.
4. Avoid unnecessary type conversions: While converting data types when necessary is essential, avoid them, as they can impact code performance. For example, only convert a string to an integer and back to a string if you don't need to perform mathematical operations on the value.
5. Use built-in Python functions: Python has many built-in functions that can be used to manipulate and work with data types. Familiarize yourself with these functions to simplify and make your code more efficient.
6. Be aware of type casting and type coercion: Type casting refers to manually 6.converting a data type, while type coercion refers to the automatic conversion of data types in certain situations. Be aware of these concepts and their impact on your code.
7. Document your code: Document your code to make it easier for others to understand your variable and data types. Use comments to explain your code and document your data types.

###**Shortcomings**

1. Limited precision for floating-point numbers: Floating-point numbers in Python have limited accuracy, so they may only sometimes be accurate for complex mathematical operations. This can lead to rounding errors and other inaccuracies.
2. Limited performance for large data sets: While Python is an efficient language, it may not perform as well as other languages when working with large data sets. This is because Python is an interpreted language and does not compile code to machine code.
3. Limited support for parallel processing: While Python has built-in support for concurrency and parallel processing, it may not be as efficient as other languages. This can impact the performance of programs that require heavy parallel processing.
4. Limited support for low-level programming: Python is a high-level language, which may not be suitable for low-level programming tasks, such as operating system development or embedded systems programming.
5. Limited control over memory allocation: Python uses automatic memory management, meaning you don't have direct control over memory allocation. This can be a disadvantage in specific applications where fine-grained control over memory allocation is necessary.
